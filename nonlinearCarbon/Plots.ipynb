```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams["figure.figsize"] = (8,5)
mpl.rcParams["savefig.bbox"] = "tight"
```

```python
cearth = 35.
tauc   = 6603.

base = np.load(f"pulse_new_{cearth}_{tauc}_{0}.npy")

Pulse = [100, 500, 1000]

# fix, ax = plt.subplots(3,1)
for pp in Pulse:
    res = np.load(f"pulse_new_{cearth}_{tauc}_{int(pp/2.13)}.npy")
    plt.plot(res[0], res[1] - base[1], label="{} GtC pulse".format(pp))
plt.legend()
plt.xlabel("Years")
plt.title("Impulse response\n heat capacity = {}, characteristic time = {}".format(cearth, tauc))
# plt.show()

plt.savefig(f"IR_{cearth}_{tauc}.pdf")
```
